```python
# create a dataframe from the crowdfunding_info sheet
crowdfunding_info_df = pd.read_excel(crowdfunding_data, sheet_name = 'crowdfunding_info')
crowdfunding_info_df.head(10)
```

```python
# Get the unique values inthe category and subcategory column.
crowdfunding_info_df['category & sub-category'].unique()
```

```python
#split the category and subcategory columns. Assign the values to df columns
crowdfunding_info_df[["category","subcategory"]] = crowdfunding_info_df["category & sub-category"].str.split('/', n=1, expand = True) #n=1 is number of splits. expand= true splits the string to seperate columns
crowdfunding_info_df.head(10)
```

```python
# Create a unique identification number for each category and subcategory in it's respective column
# Getthe distinct values in the category and subcategory columns
print(crowdfunding_info_df['category'].nunique())
print(crowdfunding_info_df['subcategory'].nunique())
```

```python
# Get the unique categories and subcategories in separate lists. 
categories = crowdfunding_info_df['category'].unique()
subcategories = crowdfunding_info_df['subcategory'].unique()

print(categories)
print()
print(subcategories)
```

```python
#Create unique numbers for the categories and subcategories

# Import numpy
import numpy as np

#Create numpy arrays that have 1-10 for the category_ids and 1-25 for the subcategory_ids.
category_ids = np.arange(1,10)
subcategory_ids = np.arange(1, 25)

print(category_ids)
print(subcategory_ids)
```

```python
# Use list comprehension to add 'cat0' to each category_id
cat_ids = ['cat0' + str(cat_id) for cat_id in category_ids]

# Use list comprehension to add "scat0" to each subcategory_id.
scat_ids = ['scat0' + str(scat_id) for scat_id in subcategory_ids]

print(cat_ids)
print(scat_ids)
```

```python
# Create category_df and subcategory_df dataframes

# Create a category dataframe 
category_df = pd.DataFrame({
    'category_id': cat_ids,
    'category': categories
})

# Create a subcategory dataframe
subcategory_df = pd.DataFrame({
    "subcategory_id":scat_ids,
    "subcategory" : subcategories
})

print(category_df)
print(subcategory_df)
```

```python
# Create a copy of the crowdfunding_info_df Dataframe as "campaign_df"
campaign_df = crowdfunding_info_df.copy()
campaign_df.head()
```

```python
# Rename the 'blurb' launched_at' and 'deadline' columns
campaign_df = campaign_df.rename(columns={'blurb':'description', 'launched_at':'launch_date', 'deadline':'end_date'})
campaign_df
```

```python
# Change the goal and pledged columns from int64 to float
campaign_df[['goal', 'pledged']] = campaign_df[['goal','pledged']].astype(float)
campaign_df.head()
```

```python
# Format the launched_date and end_date columns to datetime format
from datetime import datetime as dt

campaign_df['launch_date']= pd.to_datetime(campaign_df['launch_date'], unit='s').dt.strftime('%Y-%m-%d')
campaign_df["end_date"] = pd.to_datetime(campaign_df["end_date"], unit='s').dt.strftime('%Y-%m-%d')
campaign_df.head()
```
